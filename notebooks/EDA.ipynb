# Análise Exploratória de Dados - Anúncios de Carros nos EUA

```python
import pandas as pd
import plotly_express as px

df = pd.read_csv('../vehicles_us.csv')
df.head()
```

## Informações gerais do dataset

```python
print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
```

## Histograma de quilometragem

```python
fig = px.histogram(df, x='odometer', title='Distribuição de Quilometragem')
fig.show()
```

## Gráfico de dispersão: Preço vs Quilometragem

```python
fig = px.scatter(df, x='odometer', y='price', title='Preço vs Quilometragem')
fig.show()
```
